#### 야구 (홈런 예측)
- bip_id: 플레이 중인 공의 고유 식별자
- game_date: 게임 날짜(YYYY-MM-DD)
- home_team : 홈팀 약어  
- away_team : 원정팀 약어  
- batter_team : 타자의 팀 약어  
- batter_name : 타자의 이름  
- pitcher_name : 투수의 이름  
- Batter_id: 타자의 고유 식별자
- pitcher_id : 투수의 고유식별자
- is_batter_lefty: 왼손 타자 여부
- is_pitcher_lefty: 왼손 투수 여부
- bb_type : 타구 종류 분류
- bearing: 배트를 떠나는 공의 수평 방향 분류(예: '왼쪽' 공이 필드의 왼쪽으로 이동함)
- pitch_name: 던진 투구 유형의 이름
- park: 야구장의 고유 식별자
- inning : 경기 내 이닝 수
- outs_when_up: 현재 아웃 개수
- balls: 현재 공의 수
- strikes: 현재 파업 횟수
- plate_x: 센터 플레이트(발)의 왼쪽(-) 또는 오른쪽(+) 볼 위치
- plate_z: 홈 플레이트 위의 공 위치(피트)
- pitch_mph: 투구의 속도(시간당 마일)
- launch_speed: 공이 배트를 떠나는 속도(시간당 마일)
- launch_angle: 공이 배트를 떠나는 수직 각도(수평에 대한 각도)
- is_home_run: 홈런 여부

```python
import pandas as pd

b_df = pd.read_csv('./datasets/baseball.csv', low_memory=False)
b_df
```

```python
columns = ['bip_id', 'game_date', 'batter_name', 'pitcher_name', 'batter_id', 'pitcher_id']

b_df.drop(labels=columns, axis=1, inplace=True)
```

```python
b_df['launch_speed'] = b_df['launch_speed'].fillna(b_df['launch_speed'].mean())
b_df['launch_angle'] = b_df['launch_angle'].fillna(b_df['launch_angle'].mean())
```

```python
b_df.dropna(subset=['bb_type'], inplace=True)
```
