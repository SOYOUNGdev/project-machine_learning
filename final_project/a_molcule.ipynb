### 원자 수준 시뮬레이션 데이터베이스에서 분자 특성 예측
- 1275개의 분자 특징을 회귀분석하여, 원자화 에너지를 예측한다. (고차원 데이터 - PCA 차원축소)
- 첫 번째 열(X1): 데이터 인덱스
- 처음 1275개의 열: 분자 특징으로 작용하는 쿨롱 행렬의 항목
- 1276번째 열: 분자 구조가 얻어지는 Pubchem Id
- 1277번째 열: Quantum Espresso 패키지를 사용한 시뮬레이션을 통해 계산된 원자화 에너지

```python
import pandas as pd

r_df = pd.read_csv('./datasets/roboBohr.csv', low_memory=False)
r_df
```

```python
pre_r_df = r_df.drop(labels=['Unnamed: 0', 'pubchem_id'], axis=1)
pre_r_df
```

```python
pre_r_df.info()
```

```python
pre_r_df.isna().sum().sum()
```

```python
pre_r_df.duplicated().sum()
```

```python
pre_r_df.drop_duplicates(inplace=True)
pre_r_df = pre_r_df.reset_index(drop=True)
pre_r_df
```

```python
from sklearn.model_selection import train_test_split

features, targets = pre_r_df.iloc[:, :-1], pre_r_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)
```

```python
train_df = pd.concat([X_train, y_train], axis=1).reset_index(drop=True)
test_df = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)
```

```python
train_df.isna().sum().sum()
```

```python
test_df.isna().sum().sum()
```

```python
# PCA사용하여 차원 축소 진행
from sklearn.decomposition import PCA

# n_components = ? 차원
# 2차원을 가장 먼저
pca = PCA(n_components=2)

# fit_transform(features)
# 주어진 데이터에 대해 PCA 모델을 적합화하고, 그 결과로 차원이 축소된 데이터를 반환
# 이는 차원 축소를 수행하고 데이터를 새로운 특징 공간으로 투영하는 과정을 한 번에 처리함.
pca_train = pca.fit_transform(train_df.iloc[:, :-1])
pca_test = pca.fit_transform(test_df.iloc[:, :-1])

# 훈련된 데이터의 차원 확인
print(pca_train.shape, pca_test.shape)
```

```python
# targets
# pca_train.shape는 튜플 형태의 객체를 반환 -> [0]: 행, [1]: 열
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]
# 데이터 프레임 형태로 변환
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns) 
# 해당 데이터 프레임에 원본의 타켓 데이터 연결
# loc사용하여 맞는 행만
pca_train_df.loc[:, 'target'] = train_df['Eat']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA

features, targets = pre_r_df.iloc[:, :-1], pre_r_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# 선형 회귀 모델 l_r을 초기화
l_r = LinearRegression()
# 파이프라인 구축
# 'std': StandardScaler를 사용하여 특성을 표준화하는 단계
# 'pca': PCA를 사용하여 차원 축소하는 단계
# 'l_r': 선형 회귀 모델을 적용하는 단계
# 각 단계는 튜플 형태로 지정되며, 첫 번째 요소는 해당 변환기 또는 모델의 이름(임의의 문자열), 두 번째 요소는 해당 변환기 또는 모델 객체
pipe = Pipeline([('std', StandardScaler()), ('pca', PCA(n_components=3)), ('l_r', l_r)])
# 훈련
pipe.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    # MSLE = neg_mean_squared_log_error(y_test, prediction)
    # RMSLE = np.sqrt(neg_mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, MSLE, RMSLE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, R2))
```

```python
prediction = pipe.predict(X_test)
get_evaluation(y_test, prediction)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pipe.fit(X_train, y_train)
prediction = pipe.predict(X_train)
get_evaluation(y_train, prediction)

sns.set(style='whitegrid')
fig, ax = plt.subplots(ncols=1, figsize=(8, 4))
ax.set_title('Actual vs Predicted')
sns.scatterplot(x=y_train, y=np.ravel(prediction), ax=ax, hue=np.ravel(prediction))
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], linestyle='--', color='red')  # 대각선 추가
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```
