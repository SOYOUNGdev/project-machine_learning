```python
import pandas as pd

r_df = pd.read_csv('./datasets/roboBohr.csv', low_memory=False)
r_df
```

```python
pre_r_df = r_df.drop(labels=['Unnamed: 0', 'pubchem_id'], axis=1)
pre_r_df
```

```python
pre_r_df.drop_duplicates(inplace=True)
pre_r_df = pre_r_df.reset_index(drop=True)
pre_r_df
```

```python
from sklearn.model_selection import train_test_split

features, targets = pre_r_df.iloc[:, :-1], pre_r_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)
```

```python
train_df = pd.concat([X_train, y_train], axis=1).reset_index(drop=True)
test_df = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

# (features)
pca_train = pca.fit_transform(train_df.iloc[:, :-1])
pca_test = pca.fit_transform(test_df.iloc[:, :-1])

print(pca_train.shape, pca_test.shape)
```

```python
# targets
pca_columns = [f'pca{i + 1}' for i in range(pca_train.shape[1])]
pca_train_df = pd.DataFrame(pca_train, columns=pca_columns) 
pca_train_df.loc[:, 'target'] = train_df['Eat']

pca_columns = [f'pca{i + 1}' for i in range(pca_test.shape[1])]
pca_test_df = pd.DataFrame(pca_test, columns=pca_columns) 
pca_test_df.loc[:, 'target'] = test_df['Eat']
```

```python
from sklearn.decomposition import PCA
import numpy as np
import torch
from torch.optim import SGD
from sklearn.model_selection import train_test_split

torch.manual_seed(124)

# 데이터 불러오기 및 PCA 적용
pca = PCA(n_components=3)

# 훈련 데이터에 PCA 적용
pca_train = pca.fit_transform(train_df.iloc[:, :-1])
X_train = torch.FloatTensor(pca_train)
y_train = torch.FloatTensor(train_df['Eat'].values).view(-1, 1)

# 테스트 데이터에 PCA 적용
pca_test = pca.transform(test_df.iloc[:, :-1])
X_test = torch.FloatTensor(pca_test)
y_test = torch.FloatTensor(test_df['Eat'].values).view(-1, 1)

# 모델 초기화
W = torch.zeros((3, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = SGD([W, b], lr=1e-5)

loss_history_train = []
loss_history_test = []

epochs = 900000

# 모델 학습
for epoch in range(1, epochs + 1):
    H_train = X_train.matmul(W) + b
    loss_train = torch.mean((y_train - H_train) ** 2)

    optimizer.zero_grad()
    loss_train.backward()
    optimizer.step()

    if epoch % 100000 == 0:
        # 훈련 데이터에 대한 손실 값 저장
        loss_history_train.append(loss_train.item())

        # 검증 데이터에 대한 손실 값 계산
        with torch.no_grad():
            H_test = X_test.matmul(W) + b
            loss_test = torch.mean((y_test - H_test) ** 2)
            # 검증 데이터에 대한 손실 값 저장
            loss_history_test.append(loss_test.item())

        print('{:4d}/{}: W1: {:.4f}, W2: {:.4f}, W3: {:.4f}, b: {:.4f}, train_loss: {:.4f}, test_loss: {:.4f}'\
              .format(epoch, epochs, W[0].item(), W[1].item(), W[2].item(), b.item(), loss_train.item(), loss_test.item()))
```

```python
import matplotlib.pyplot as plt

# 훈련 및 테스트 손실 그래프 그리기
plt.plot(range(100000, epochs + 1, 100000), loss_history_train, label='Train Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Train Loss over Epochs')
plt.legend()
plt.show()

plt.plot(range(100000, epochs + 1, 100000), loss_history_test, label='Test Loss', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Test Loss over Epochs')
plt.legend()
plt.show()
```
