### 부동산 가격 예측
- 부동산 특성을 회귀분석하여, 부동산 가격을 예측한다. (비선형)

1. transaction date: 거래 날짜
2. house age: 현재년도 - 준공년도
3. distance to the nearest MRT station: 역까지의 거리
4. number of convenience stores: 주변 편의점 수
5. latitude: 위도
6. longitude: 경도
7. house price of unit area: 단위 면적 당 가격

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

features, targets = pre_r_df.iloc[:, :-1], pre_r_df.iloc[:, -1]

# poly_features = PolynomialFeatures(degree=2).fit_transform(features)

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```

#### linear regression 사용, R2 score를 올려보기 위해 다항회귀 사용

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

features, targets = pre_r_df.iloc[:, :-1], pre_r_df.iloc[:, -1]

poly_features = PolynomialFeatures(degree=2).fit_transform(features)

X_train, X_test, y_train, y_test = \
train_test_split(poly_features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```

### 비선형

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

features, targets = pre_r_df.iloc[:, :-1], pre_r_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor()
lgb_r = LGBMRegressor(n_estimators=100)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]
for model in models:
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    print(model.__class__.__name__)
    get_evaluation(y_test, prediction)
```

```python
import matplotlib.pyplot as plt

rf_r.fit(X_train, y_train)
prediction = rf_r.predict(X_train)
get_evaluation(y_train, prediction)

sns.set(style='whitegrid')
fig, ax = plt.subplots(ncols=1, figsize=(8, 4))
ax.set_title('Actual vs Predicted')
sns.scatterplot(x=y_train, y=np.ravel(prediction), ax=ax, hue=np.ravel(prediction))
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], linestyle='--', color='red')  # 대각선 추가
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
import matplotlib.pyplot as plt

rf_r.fit(X_train, y_train)
prediction = rf_r.predict(X_test)
get_evaluation(y_test, prediction)

sns.set(style='whitegrid')
fig, ax = plt.subplots(ncols=1, figsize=(8, 4))
ax.set_title('Actual vs Predicted')
sns.scatterplot(x=y_test, y=np.ravel(prediction), ax=ax, hue=np.ravel(prediction))
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')  # 대각선 추가
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
from sklearn.model_selection import KFold, cross_val_score
import numpy as np

kfold = KFold(n_splits=10, shuffle=True)

scores = cross_val_score(rf_r, X_train, y_train, cv=kfold, scoring='r2')
for score in scores:
    print(f'{np.round(score, 4)}')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train_train, X_val_train, X_train_train, y_val_test = \
train_test_split(X_train, y_train, test_size=0.2, random_state=124)

# 교차검증 수행
cv_scores_train = cross_val_score(rf_r, X_train, y_train, cv=10, scoring='r2')
cv_scores_test = cross_val_score(rf_r, X_test, y_test, cv=10, scoring='r2')

# 훈련 세트와 테스트 세트 간의 성능 차이 계산
mean_train_score = cv_scores_train.mean()
mean_test_score = cv_scores_test.mean()

print('훈련 데이터세트의 r2평균 :', mean_train_score)
print('검증 데이터세트의 r2평균 :', mean_test_score)
```
