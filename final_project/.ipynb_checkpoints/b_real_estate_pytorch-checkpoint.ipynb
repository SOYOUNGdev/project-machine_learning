```python
import pandas as pd

r_df = pd.read_csv('./datasets/Real estate.csv')
r_df
```

```python
pre_r_df = r_df.drop(labels=['No'], axis=1)
pre_r_df
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
result = std.fit_transform(pre_r_df)
std_pre_r_df = pd.DataFrame(result, columns=pre_r_df.columns)
std_pre_r_df
```

```python
condition = True

for column in std_pre_r_df.columns:
    condition &= std_pre_r_df[column].between(-1.96, 1.96)

std_pre_r_df = std_pre_r_df[condition]
std_pre_r_df
```

```python
pre_r_df = pre_r_df.iloc[std_pre_r_df.index].reset_index(drop=True)
pre_r_df
```

```python
pre_r_df.corr()['Y house price of unit area'].sort_values(ascending=False)[1:]
```

```python
pre_r_df = pre_r_df.drop(labels=['X1 transaction date', 'X6 longitude'], axis=1)
pre_r_df
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
import numpy as np
import torch
from torch.optim import SGD
from sklearn.model_selection import train_test_split

torch.manual_seed(124)

features, targets = pre_r_df.iloc[:, :-1], pre_r_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

X_train = torch.FloatTensor(X_train.values)
y_train = torch.FloatTensor(y_train.values).view(-1, 1)

X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values).view(-1, 1)

W = torch.zeros((4, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = SGD([W, b], lr=1e-7)

loss_history_train = []
loss_history_test = []

epochs = 1000000

for epoch in range(1, epochs + 1):
    H_train = X_train.matmul(W) + b
    loss_train = torch.mean((y_train - H_train) ** 2)

    optimizer.zero_grad()
    loss_train.backward()
    optimizer.step()

    if epoch % 100000 == 0:
        if(epoch > 100000):
            # 훈련 데이터에 대한 손실 값 저장
            loss_history_train.append(loss_train.item())

        # 검증 데이터에 대한 손실 값 계산
        with torch.no_grad():
            H_test = X_test.matmul(W) + b
            loss_test = torch.mean((y_test - H_test) ** 2)
            if(epoch > 100000):
                # 검증 데이터에 대한 손실 값 저장
                loss_history_test.append(loss_test.item())

        print('{:4d}/{}: W1: {:.4f}, W2: {:.4f}, W3: {:.4f}, b: {:.4f}, train_loss: {:.4f}, test_loss: {:.4f}'\
              .format(epoch, epochs, W[0].item(), W[1].item(), W[2].item(), b.item(), loss_train.item(), loss_test.item()))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 새로운 Figure 객체 생성 및 크기 설정
plt.figure(figsize=(6, 5))

# 훈련 및 검증 데이터의 손실 값의 개수를 x축으로 사용
x_train = list(range(1, len(loss_history_train) + 1))
x_val = list(range(1, len(loss_history_test) + 1))

# 훈련 데이터와 검증 데이터의 손실 값을 그래프로 표시
plt.plot(x_train, loss_history_train, label='Train Loss')
plt.plot(x_val, loss_history_test, label='Validation Loss')

# 검증 손실이 감소하는 지점을 찾아 빨간 선으로 표시
min_val_loss_index = loss_history_test.index(min(loss_history_test))
min_val_loss_epoch = x_val[min_val_loss_index]
plt.axvline(min_val_loss_epoch, color='red', linestyle='--', label='Min Validation Loss')

# 최소 검증 손실의 바로 이전 epoch와 손실 값을 찾아서 그래프에 표시
prev_epoch = min_val_loss_epoch - 1
prev_val_loss = loss_history_test[min_val_loss_index - 1]
plt.plot(prev_epoch, prev_val_loss, marker='o', markersize=5, color='orange', label='Previous Validation Loss')

# 이전 epoch와 손실 수치를 텍스트로 추가 (상단에 표시)
plt.text(prev_epoch, prev_val_loss, f'Epoch:{(prev_epoch+1) * 100000}\nLoss: {prev_val_loss:.4f}', 
         verticalalignment='top', horizontalalignment='right', color='blue')

# 최소 검증 손실의 바로 이후 epoch와 손실 값을 찾아서 그래프에 표시
next_epoch = min_val_loss_epoch + 1
next_val_loss = loss_history_test[min_val_loss_index + 1]
plt.plot(next_epoch, next_val_loss, marker='o', markersize=5, color='green', label='Next Validation Loss')

# 이후 epoch와 손실 수치를 텍스트로 추가 (상단에 표시)
plt.text(next_epoch, next_val_loss, f'Epoch:{(next_epoch+1) * 100000}\nLoss: {next_val_loss:.4f}', 
         verticalalignment='top', horizontalalignment='left', color='green')

# 최소 검증 손실 위치에 텍스트 추가
plt.text(min_val_loss_epoch, min(loss_history_test), f'Epoch:{(min_val_loss_epoch+1) * 100000}\nLoss: {min(loss_history_test):.4f}', 
         verticalalignment='bottom', horizontalalignment='left', color='red')

# 그래프 레이블 및 타이틀 추가
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')

# 범례 추가
plt.legend()

# 그래프 출력
plt.show()
```

```python
import torch
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error, r2_score

# 모델 불러오기
checkpoint = torch.load('best_model.pth')
W = checkpoint['W']
b = checkpoint['b']

# 예측값 계산
H_train = torch.matmul(X_train, W) + b
H_test = torch.matmul(X_test, W) + b

# MSE (Mean Squared Error)
mse_train = mean_squared_error(y_train.detach().numpy(), H_train.detach().numpy())
mse_test = mean_squared_error(y_test.detach().numpy(), H_test.detach().numpy())
print("MSE :")
print("훈련 세트:", mse_train)
print("테스트 세트:", mse_test)

# RMSE (Root Mean Squared Error)
rmse_train = mean_squared_error(y_train.detach().numpy(), H_train.detach().numpy(), squared=False)
rmse_test = mean_squared_error(y_test.detach().numpy(), H_test.detach().numpy(), squared=False)
print("\nRMSE :")
print("훈련 세트:", rmse_train)
print("테스트 세트:", rmse_test)

# MSLE (Mean Squared Log Error)
msle_train = mean_squared_log_error(y_train.detach().numpy(), H_train.detach().numpy())
msle_test = mean_squared_log_error(y_test.detach().numpy(), H_test.detach().numpy())
print("\nMSLE :")
print("훈련 세트:", msle_train)
print("테스트 세트:", msle_test)

# RMSLE (Root Mean Squared Log Error)
rmsle_train = mean_squared_log_error(y_train.detach().numpy(), H_train.detach().numpy(), squared=False)
rmsle_test = mean_squared_log_error(y_test.detach().numpy(), H_test.detach().numpy(), squared=False)
print("\nRMSLE :")
print("훈련 세트:", rmsle_train)
print("테스트 세트:", rmsle_test)

# R-squared (R2) score
r2_train = r2_score(y_train.detach().numpy(), H_train.detach().numpy())
r2_test = r2_score(y_test.detach().numpy(), H_test.detach().numpy())
print("\nR2 :")
print("훈련 세트:", r2_train)
print("테스트 세트:", r2_test)
```
