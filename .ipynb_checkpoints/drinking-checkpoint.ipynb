### 흡연 및 음주 데이터 세트
- sex: 남성 여성
- age: 5년으로 반올림
- height: 5cm로 반올림
- weight: 몸무게
- waistline: 
- sight_left: 시력(좌)
- sight_right: 시력(우)
- hear_left: 청각(좌) -> 1: 정상, 2: 비정상
- hear_right: 청각(우) -> 1: 정상, 2: 비정상
- SBP: 수축기 혈압
- DBP: 확장기 혈압
- BLDS: 공복혈당
- tot_chole: 총 콜레스테롤
- HDL_chole: HDL
- LDL_chole: LDL
- triglyceride: 중성지방
- hemoglobin: 헤모글로빈
- urine_protein: 소변 내 단백질 1(-), 2(+/-), 3(+1), 4(+2), 5(+3), 6(+4)
- serum_creatinine: 혈청 크레아티닌
- SGOT_AST: AST
- SGOT_ALT: SLT
- gamma_GTP: GTP
- SMK_stat_type_cd: 흡연상태 -> 1: 안함, 2: 끊음, 3: 흡연
- DRK_YN: 음주 유무

```python
pre_s_df = s_df.drop_duplicates()
pre_s_df = pre_s_df.reset_index(drop=True)
pre_s_df
```

```python
# 수치형이 아닌 범주형 feature들 삭제
pre_s_df = pre_s_df.drop(labels=['sex', 'SMK_stat_type_cd', 'urine_protein', 'hear_right', 'hear_left'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
result = encoder.fit_transform(pre_s_df['DRK_YN'])
pre_s_df['DRK_YN'] = result
```

```python
from statsmodels.api import OLS

features, targets = pre_s_df.iloc[:, :-1], pre_s_df.iloc[:, -1]

model = OLS(targets, features)
print(model.fit().summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    # MSLE = mean_squared_log_error(y_test, prediction)
    # RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, MSLE, RMSLE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, R2))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = pre_s_df.iloc[:, :-1], pre_s_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.5, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train.values, y_train.values)
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```

```python
from sklearn.preprocessing import PolynomialFeatures

features, targets = pre_s_df.iloc[:, :-1], pre_s_df.iloc[:, -1]

poly_features = PolynomialFeatures(degree=2).fit_transform(features)

X_train, X_test, y_train, y_test = \
train_test_split(poly_features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    # MSLE = mean_squared_log_error(y_test, prediction)
    # RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    # print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
    #       .format(MSE, RMSE, MSLE, RMSLE, R2))
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, R2))
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

features, targets = pre_s_df.iloc[:, :-1], pre_s_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor()
lgb_r = LGBMRegressor(n_estimators=100)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]
for model in models:
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    print(model.__class__.__name__)
    get_evaluation(y_test, prediction)
```
